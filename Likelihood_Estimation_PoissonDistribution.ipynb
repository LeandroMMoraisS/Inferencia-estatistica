```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# Geração de dados
mu_true = 4.5  # Valor verdadeiro do parâmetro mu
N = 50  # Número de dados
data = np.random.poisson(mu_true, N)  # Dados Poisson
```

```python
data
```

```python
# Cálculo da curva de verossimilhança
mu_vals = np.linspace(0, 10, 1000)  # Valores de mu para avaliação
log_likelihood = np.zeros_like(mu_vals)  # Inicialização da curva de verossimilhança
for i, mu in enumerate(mu_vals):
    log_likelihood[i] = np.sum(poisson.logpmf(data, mu))  # Cálculo do logaritmo da função de verossimilhança
```

```python
# Plotagem da curva de verossimilhança
plt.plot(mu_vals, np.exp(log_likelihood - np.max(log_likelihood)), label='Verossimilhança')
plt.axvline(mu_true, color='red', label='Valor verdadeiro de mu')
plt.xlabel('mu')
plt.ylabel('Verossimilhança')
plt.legend()
plt.show()
```
